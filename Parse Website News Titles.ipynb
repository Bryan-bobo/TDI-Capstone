```python
## UP SINGLE WORD COUNT 4%
import nltk
nltk.download('punkt')
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import nltk
from nltk.corpus import stopwords

# Download the NLTK stop words
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Define the dates to search for
dates = [datetime(2018, 4, 4), 
         datetime(2018, 12, 26),
         datetime(2020, 2, 28),
         datetime(2020, 3, 13),
         datetime(2020, 3, 17),
         datetime(2020, 3, 24),
         datetime(2020, 3, 26),
         datetime(2020, 4, 6),
         datetime(2020, 5, 14),
         datetime(2020, 6, 15),
         datetime(2021, 2, 24),
         datetime(2021, 3, 26),
         datetime(2021, 5, 19),
         datetime(2022, 1, 24),
         datetime(2022, 1, 31),
         datetime(2022, 5, 26),
         datetime(2022, 10, 13),
         datetime(2022, 10, 21),
         datetime(2022, 10, 28),
         datetime(2022, 11, 10),
         datetime(2022, 11, 30),
         datetime(2023, 2, 1),
         datetime(2023, 3, 16)
        ]

# Define a dictionary to store the word frequencies
word_freq = {}

# Loop through the dates
for date in dates:
    # Convert the date to the URL format used by TechCrunch
    url_date = date.strftime("%Y/%m/%d")

    # Set up the URL for the TechCrunch archive page for the specified date
    url = f'https://techcrunch.com/{url_date}/'

    # Send a GET request to the URL and parse the HTML content using BeautifulSoup
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the article titles
    titles = soup.select('.post-block__title a')

    # Loop through the titles
    for title in titles:
        # Get the text of the title and split it into words
        title_text = title.get_text().strip()
        words = nltk.word_tokenize(title_text)

        # Loop through the words and update the word frequency dictionary
        for word in words:
            # Convert the word to lowercase and remove any non-alphabetic characters
            word = word.lower()
            word = ''.join(filter(str.isalpha, word))

            # Exclude stop words
            if word not in stop_words and word != '':
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1

# Sort the word frequency dictionary by frequency in descending order
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Print the top 10 words by frequency
print("Top 20 words by frequency:")
for word, freq in sorted_word_freq[:20]:
    print(f"{word}: {freq}")
```

```python
## UP DOUBLE WORD COUNT 4%
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords

# set the URL and date
url = "https://techcrunch.com/"
dates = [datetime(2018, 4, 4), 
         datetime(2018, 12, 26),
         datetime(2020, 2, 28),
         datetime(2020, 3, 13),
         datetime(2020, 3, 17),
         datetime(2020, 3, 24),
         datetime(2020, 3, 26),
         datetime(2020, 4, 6),
         datetime(2020, 5, 14),
         datetime(2020, 6, 15),
         datetime(2021, 2, 24),
         datetime(2021, 3, 26),
         datetime(2021, 5, 19),
         datetime(2022, 1, 24),
         datetime(2022, 1, 31),
         datetime(2022, 5, 26),
         datetime(2022, 10, 13),
         datetime(2022, 10, 21),
         datetime(2022, 10, 28),
         datetime(2022, 11, 10),
         datetime(2022, 11, 30),
         datetime(2023, 2, 1),
         datetime(2023, 3, 16)
        ]

# set the stop words
stop_words = set(stopwords.words('english'))

# create a list to store all the titles
titles = []

# iterate through each date
for date in dates:
    date_str = date.strftime("%Y/%m/%d")
    # send a request to the URL with the date as a parameter
    r = requests.get(url + date_str)

    # create a BeautifulSoup object
    soup = BeautifulSoup(r.content, 'html.parser')

    # find all the title elements on the page
    title_elements = soup.find_all('h2', {'class': 'post-block__title'})

    # extract the text from each title element and append it to the titles list
    for title_element in title_elements:
        title = title_element.text.strip()
        titles.append(title)

# tokenize the text into bigrams
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(' '.join(titles))
bigrams = nltk.bigrams(tokens)

# filter out bigrams containing stop words
filtered_bigrams = [bigram for bigram in bigrams if bigram[0] not in stop_words and bigram[1] not in stop_words]

# count the frequency of each remaining bigram
frequency_dist = nltk.FreqDist(filtered_bigrams)

# print the most common bigrams and their frequencies
for bigram, frequency in frequency_dist.most_common(10):
    print(bigram, frequency)
```
